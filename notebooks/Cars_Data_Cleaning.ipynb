```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cars_data.csv")
df = df.rename(columns={'kolimeter': 'kilometer', 'feul_type': 'fuel_type'})
df
```

```python
df.info()
```

```python
df['city'] = df['city'].str.strip()
df.rename(columns={'model': 'model_year'}, inplace=True)
```

```python
import numpy as np

def clean_price(x):
    x = str(x).lower().replace('pkr', '').strip()  # convert to string first

    if 'crore' in x:
        return float(x.replace('crore', '').strip()) * 10000000
    if 'lac' in x:
        return float(x.replace('lacs', '').replace('lac', '').strip()) * 100000

    try:
        return float(x)  # if it's already a number
    except:
        return np.nan
```

```python
df['price'] = df['price'].apply(clean_price)
df['price'] = df['price'].fillna(9800000)
df['price'].isnull().sum()

df['price'] = df['price'].astype(int)
df.head(5)
```

```python
df['kilometer'] = df['kilometer'].str.replace(',', '').str.replace(' km', '')
df['engine'] = df['engine'].str.replace(' cc', '').str.replace(' kWh','')
df.head(5)
df['kilometer'] = pd.to_numeric(df['kilometer'], errors='coerce')
df['kilometer'] = df['kilometer'].astype('Int64')
df['engine'] = pd.to_numeric(df['engine'], errors='coerce')   
df['engine'] = df['engine'].astype('Int64')   
```

```python
df['name'] = df['name'].str.replace(r'\b(19|20)\d{2}\b', '', regex=True)
df['name'] = df['name'].str.replace(r'\bfor sale\b', '', regex=True, case=False)
df['name'] = df['name'].str.replace(r'\s+', ' ', regex=True).str.strip()
df.head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
df['city'] = df['city'].str.replace('Dera ismail khan','DI Khan').str.replace('Malakand Agency','Malakand').str.strip()
df['brand'] = df['name'].str.split().str[0]
df['model_name'] = df['name'].str.split().str[1]
df.head()
```
